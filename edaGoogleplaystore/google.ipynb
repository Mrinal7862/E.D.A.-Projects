```python
#1Mb = 1024 kb
```

```python
def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M', ''))
        res = 1024 * res
        return res

    elif str(item)[-1] == 'K':
        res = float(str(item).replace('k', ''))
        return res

    else:
        return str(np.NaN)
        
```

```python
df_copy['Size'].describe
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
